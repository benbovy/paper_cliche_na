# Analysis and visualization of the NA direct-search results

Modules used in this notebook:

```python
import os

import numpy as np
from scipy.stats import gaussian_kde
import pandas as pd

import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas.tools.plotting as pdplt
import seaborn as sns

# D3 plots with pandas objects in the notebook
# https://github.com/mlunacek/smash
import smash

%matplotlib inline
```

Name of the inversion run:

```python
na_run_name = "na_run20"
```

**Comment on `na_run20`:** *in this run, the NA parameters were chosen for exhaustive exploring of the parameter space rather than fine optimization (1024 samples randomly generated at the first iteration, then 3 samples generated in each of the 32 best-fit Voronoi cells at each next iteration).*

## Import data + pre-processing

Import inversion inputs / outputs as Python dicts or Pandas dataframes.

```python
# Inversion run directory
na_run_dir = os.path.abspath(
    os.path.join(os.pardir, na_run_name)
)

# NA inversion parameters
na_params_file = os.path.join(na_run_dir, "input", "na.in")
na_in = pd.read_table(na_params_file, sep=':', comment='#',
                      engine='python', header=None, usecols=[0])
pnames = ('algorithm', 'n_iter', 'n_spl_init', 'n_spl_iter',
          'n_cell_iter', 'seed', 'init_type', 'verbose',
          'time_mode', 'debug_mode')
ptypes = (int, int, int, int, int, str, int, int, str, str)
pvalues = (t(p) for t, p in zip(ptypes, na_in.get(0)))
na_params = dict(zip(pnames, pvalues))

# Model parameter search ranges
p_ranges_file = os.path.join(na_run_dir, "input", "p_ranges.in")
p_ranges = pd.read_csv(p_ranges_file,
                       index_col='name',
                       dtype={'log': np.bool})

# NA search samples
na_samples_file = os.path.join(na_run_dir, "output", "na_samples.out")
na_samples = pd.read_table(na_samples_file)
```

Re-calculate some intermediate variables and create derived dataframes used for visualization.

```python
# correct a bug (wrong P_0 values but didn't impact the results)
na_samples['P_0'] /= 4.0
p_ranges.loc['P_0', 'min'], p_ranges.loc['P_0', 'max'] = 2.5e-7, 2.5e-4

# re-calculate the mean of the K-S statistics that were
# computed for each soil class and topography derivative
ks_cols = [c for c in na_samples.columns if '_D_' in c]
d_mean = na_samples.loc[:, ks_cols].mean(axis=1)
na_samples.loc[:, 'D_mean'] = d_mean

# re-calculate the RSMD values from each sample proportions
sp_cols = [c for c in na_samples.columns if c.startswith('SP_')]
sp_obs = [0.03693368, 0.17780696, 0.55456336, 0.23069599]  # observed sample prop
ssd = ((sp_obs - na_samples.loc[:, sp_cols])**2).sum(axis=1)
na_samples.loc[:, 'RMSD'] = np.sqrt(ssd / len(sp_obs))

# assign the iteration number of the NA direct-search to each sample
na_iter = np.concatenate(
    [np.zeros((na_params['n_spl_init'])),
     np.repeat(np.arange(1, na_params['n_iter']),
               na_params['n_spl_iter'])]
)
na_samples.loc[:, 'NA_iter'] = na_iter[:len(na_samples)]
```

Create derived dataframes used for easier visualization and analysis.

```python
# drop columns that are not very useful anymore
na_dataset = na_samples.drop(ks_cols + sp_cols + ['mname'],
                             axis=1)

# transform values for parameters with log search scale
na_dataset_log = na_dataset.copy()
for col in na_dataset.columns[:-3]:
    if col in p_ranges.index and p_ranges['log'][col]:
        na_dataset_log.loc[:, col] = np.log(na_dataset[col])
        logname = "log({})".format(col)
        na_dataset_log.rename(columns={col: logname},
                              inplace=True)
```

Additional stuff for easier visualization.

```python
# map column names to math-formatted parameter / variables
na_labels = {
    'P_0': '$P_0$',
    'K_r12': '$K_r$',
    'K_d': '$K_d$',
    'K_dd': '$K_{dd}$',
    'K_g': '$K_g$',
    'h_init': '$h_{t0}$',
    'misfit': '$M$',
    'D_mean': '$\overline{D}*$',
    'RMSD': 'RMSD'
}
```

## Trace plots

Trace plots of the samples (models) generated in the model parameter space by the NA direct-search algorithm. In the subplots below, each dot represents a sample and the x-axis represents the iterations of the algorithm. The dots are drawn with some degree of alpha transparency so that it is easier to identify the zones of high sampling density.

```python
fig, axes = plt.subplots(3, 3, figsize=(10, 6))

for col, ax in zip(na_dataset.columns, axes.flatten()):
    if col in p_ranges.index and p_ranges.loc[col, 'log']:
        logy = True
    else:
        logy = False
    if col in ['misfit', 'D_mean', 'RMSD']:
        pt_clr = '#7A68A6'
    else:
        pt_clr = '#348ABD'
    
    trace = na_dataset.get(col)
    na_dataset.plot(kind='scatter', x='NA_iter', y=col, ax=ax,
                    logy=logy, s=50, alpha=0.15, color=pt_clr)
    
    if col in p_ranges.index:
        plt.setp(ax, ylim=p_ranges.loc[col, ['min','max']])
    plt.setp(ax, xlim=[-1, na_params['n_iter']])
    plt.setp(ax, ylabel=na_labels[col], xlabel='NA iteration')

fig.tight_layout()
```

**Comment on `na_run20`:** *We see in these trace plots remarkable convergence patterns, despite that the NA algorithm has not converged towards a unique solution. The $\overline{D}*$ values got rapidly minimized after only a few iterations, but are stuck in the [0.2, 0.3] interval (no better fit could be reached). Although it took more iterations to minimize the RMSD values, the lasts iterations don't show much evolution in both the computed $\overline{D}*$ and RMSD values. The trace of the $M$ (misfit) values, however, show a significant evolution even during the lasts iterations, but this is mostly due to the very small lowering of the computed RMSD values that were already close to 0 at the end of the process. Given our purpose, which consists of exploring the parameter space rather than performing a fine calibration of the model parameters, we can thus reasonably say that the NA direct-search has completed with an enough number of iterations.* 

## Plot of misfit vs. intermediate statistics

Plot to show the relationships between the RMSD, $\overline{D}*$ and $M$ values computed for each sample.

```python
fig, axes = plt.subplots(1, 3, figsize=(10, 3))

col_pairs = (('RMSD', 'D_mean'),
             ('RMSD', 'misfit'),
             ('D_mean', 'misfit'))

for (x, y), ax in zip(col_pairs, axes):
    na_dataset.plot(kind='scatter', x=x, y=y, ax=ax, alpha=0.2)
    plt.setp(ax, xlabel=na_labels[x], ylabel=na_labels[y])
    
fig.tight_layout()
```

**Comment on `na_run20`:** *We see a consistent, positive - though non-linear - relationship between RMSD and $\overline{D}*$ values (except for some samples with low $\overline{D}*$ values < 2 and RMSD values between 0.2 and 0.3). The three plots together show that RMSD mostly contribute to the variability of $M$ for $M$ values below -4. This explains the high sensitivity of $M$ to small differences in low RMSD values.*   

## Sample distribution plots

Sample distributions allow to see, for each model parameter, the regions of best fit (i.e., the regions that have been preferably sampled by the NA direct-search algorithm as the iterations proceeded). The plots below show histograms and Kernel densisty Estimation (KDE) for each model parameter. The plotted distributions don't include the samples generated randomly at the first iteration of the NA direct-search algorithm.

```python
fig, axes = plt.subplots(2, 3, figsize=(8, 5))

for col, ax in zip(na_dataset.columns, axes.flatten()):
    
    # don't include samples from 1st NA iteration
    col_data = na_dataset[na_params['n_spl_init']:].get(col)
    
    xlim = p_ranges.loc[col, ['min', 'max']].values.astype('f')
    
    # calculate KDE on log or linear scale
    if p_ranges.loc[col, 'log'] == True: 
        bins = np.logspace(*np.log10(xlim), num=20)
        kde_x = np.logspace(*np.log10(xlim), num=200)
        kde_gen = gaussian_kde(np.log10(col_data))
        kde_y = kde_gen.evaluate(np.log10(kde_x))
        plt.setp(ax, xscale='log')
    else:
        bins = np.linspace(*xlim, num=20)
        kde_x = np.linspace(*xlim, num=200)
        kde_gen = gaussian_kde(col_data)
        kde_y = kde_gen.evaluate(kde_x)
    
    sns.distplot(col_data, bins=bins, kde=False, hist=True, ax=ax)
    ax2 = ax.twinx()
    ax2.plot(kde_x, kde_y)
    
    plt.setp(ax, yticks=[], xlim=xlim, xlabel=na_labels[col])
    plt.setp(ax2, yticks=[])

fig.tight_layout()
```

## Scatterplot matrix



```python
# sort and subset data
na_sorted = na_dataset.sort('misfit', ascending=False)
na_subset = na_sorted.loc[:, 'P_0':'h_init']
n_diag = len(na_subset.columns)

# color map
cmap = sns.cubehelix_palette(start=.5, rot=-.75,
                             as_cmap=True, reverse=True)

# scatter matrix
axes = pdplt.scatter_matrix(na_subset, c=na_sorted['misfit'],
                            s=50, cmap=cmap, edgecolors='none',
                            alpha=0.4, figsize=(11, 12), grid=False)
fig = axes[0][0].figure

# colorbar
dummy_scatter = axes[0][0].scatter(
    na_sorted['misfit'], na_sorted['misfit'],
    c=na_sorted['misfit'], cmap=cmap
)
cax = fig.add_axes([0.2, 0.05, 0.6, 0.02])
clb = plt.colorbar(dummy_scatter, cax=cax,
                   orientation='horizontal')
clb.set_label(na_labels['misfit'])

# clear diagonal subplots
for i in range(n_diag):
    plt.setp(axes[i][i], axis_bgcolor='w')
    axes[i][i].clear()

# axis limits (and log-scale if needed)
for i in range(n_diag):
    icol = na_subset.columns[i]
    ylim = p_ranges.loc[icol, ['min', 'max']]
    for j in range(n_diag):
        jcol = na_subset.columns[j]
        xlim = p_ranges.loc[jcol, ['min', 'max']]
        plt.setp(axes[i][j], xlim=xlim, ylim=ylim)
        if p_ranges.loc[icol, 'log']:
            plt.setp(axes[i][j], yscale='log')
        if p_ranges.loc[jcol, 'log']:
            plt.setp(axes[i][j], xscale='log')

# axis labels
for i in range(n_diag):
    icol = na_subset.columns[i]
    label = na_labels[icol]
    plt.setp(axes[-1][i], xlabel=label)
    plt.setp(axes[i][0], ylabel=label)

# subplots spacing
fig.subplots_adjust(wspace=0.02, hspace=0.02, bottom=0.13)
```
